### Step 4. How many products cost more than $10.00?

```python
# need to replace '$' as float instead of integer so that the cent value would remain as cents
# use chipo['item_price'] = chipo['item_price'].str.replace('[\$\,]', '', regex=True).astype(float)
```

```python
chipo =  pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep = "\t")
chipo['item_price'] = chipo['item_price'].str.replace('[\$\,]', '', regex=True).astype(float)
chipo
```

```python
# drop duplicates in item_name and quantity, transform data 
# newdf = df1.drop_duplicates(  subset = ['order_id', 'customer_id'],  keep = 'last').reset_index(drop = True)
chipo = chipo.drop_duplicates(subset = ['item_name', 'quantity'],   keep = 'last').reset_index(drop = True)
chipo
```

```python
# select only the products with quantity equals to 1
# How many products cost more than $10
# len(df[(df['col1']=='value1') & (df['col2']=='value2')])

len(chipo[(chipo['quantity']==1) & (chipo['item_price']>10.00)])
```

### Step 5. What is the price of each item? 
##### print a data frame with only two columns item_name and item_price

```python
price_per_item = chipo[['item_name', 'item_price']]
price_per_item
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by = "item_name")
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by = "item_price", ascending = False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# mode function
chipo_salad = chipo[chipo.item_name == "Veggie Salad Bowl"]

len(chipo_salad)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo_cannedsoda = chipo[chipo.item_name == "Canned Soda"]

len(chipo_cannedsoda)
```
